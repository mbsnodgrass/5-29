```python
#Simpson diversity index function

def simpson_diversity(row):
    groups = row[['B02001_002E', 'B02001_003E', 'B02001_004E', 'B02001_005E', 'B02001_006E']]
    N = groups.sum()
    if N <= 1:
        return 0
    numerator = sum(n * (n - 1) for n in groups)
    denominator = N * (N - 1)
    D = 1 - (numerator / denominator)
    return D

# Apply to DataFrame
race['Diversity_Score'] = race.apply(simpson_diversity, axis=1)
```

```python
from rapidfuzz import process


#Fuzzy merge
```

```python
census_cities = census_150['NAME'].unique()
zillow_cities = rent['NAME'].unique()
```

```python
city_match = {}
for zc in zillow_cities:
    match = process.extractOne(zc, census_cities, score_cutoff=90)
    city_match[zc] = match[0] if match else None
```

```python
rent['Matched_Name'] = rent['NAME'].map(city_match)
```
